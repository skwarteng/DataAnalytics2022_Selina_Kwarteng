```python
new_df = merged_df.sample(n = 50000)
 
#output 
new_df
```

```python
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#feature selection

X = new_df.drop(columns = ['Value_x', 'Value_y'])
y = new_df[['Value_x', 'Value_y']]
```

```python
#one-hot encoding the features
X = pd.get_dummies(X)
```

```python
y.Value_y = y.Value_y.apply(lambda x : x.replace(',',''))
y.Value_y = y.Value_y.astype(float)
```

```python
y.dtypes
```

```python
X.astype("int")
```

```python
y['Value_x'] = y['Value_x'].astype(int)
y['Value_y'] = y['Value_y'].astype(int)
```

```python
#decision tree classifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

dt1 = DecisionTreeClassifier(max_depth=4, random_state=10)
dt1.fit(X, y['Value_x'])
valuex_ = np.mean(cross_val_score(dt1, X, y['Value_x'], cv=5))

dt2 = DecisionTreeClassifier(max_depth=4, random_state=10)
dt2.fit(X, y['Value_y'])
valuey_ = np.mean(cross_val_score(dt2, X, y['Value_y'], cv=5))

print(valuex_)
print(valuey_)
```

```python
y['Value_y'] = y["Value_y"].astype(float)
```

```python
#split the train and test dataset

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#create functions to perform algorithms

def predict_valuex(regression_algo):
    r = regression_algo()
    r.fit(X_train,y_train['Value_x'])
    y_pred = r.predict(X_test)
    rSquare = r2_score(y_test['Value_x'],y_pred)
    valuex.append(rSquare)

def predict_valuey(algos):
    r = algos()
    r.fit(X_train,y_train['Value_y'])
    y_pred = r.predict(X_test)
    rSquare = r2_score(y_test['Value_y'],y_pred)
    valuey.append(rSquare)
    
models = [RandomForestRegressor, Lasso]

valuex = []
valuey = []

for i in models:
  predict_valuex(i)

for j in models:
  predict_valuey(j)

print(valuex)
print(valuey)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y['Value_y'])

print(round(np.mean(cross_val_score(knn, X, y['Value_y'], cv=5)), 2))
```

```python
x_vals=[]
y_vals=[]
for k in range(3, 11):
  x_vals.append(k)
  knn = KNeighborsClassifier(n_neighbors=k)
  y_vals.append(round(np.mean(cross_val_score(knn, X, y['Value_x'], cv=5)), 2))

print(x_vals)
print(y_vals)
```
